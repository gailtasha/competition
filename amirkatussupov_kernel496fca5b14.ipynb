```python
X = train.iloc[:, 2:].values
Y = train.iloc[:, [1]].values
X
```

```python
knn.fit(X_train, Y_train)
lg.fit(X_train, Y_train)
gnb.fit(X_train, Y_train)
dt.fit(X_train, Y_train)
rfc.fit(X_train, Y_train)
```
